**Line Plot**

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
```

```python
x
```

```python
y = x*2
```

```python
y
```

```python
plt.plot(x,y)
plt.show()
```

**Adding Title and Labels to Plot**

```python
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

**Changing Line Aesthetics**

```python
plt.plot(x,y, color='g',linestyle=':',linewidth=2)
plt.show()
```

```python
y1=2*x
```

```python
y2=3*x
```

```python
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.plot(x,y2,color='r',linestyle='-.',linewidth=3)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

**Sub-Plot**

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color='r',linestyle='-',linewidth=3)

plt.subplot(1,2,2)
plt.plot(x,y2,color='b',linestyle=':',linewidth=2)

plt.grid(True)

plt.show()
```

**Bar-Plot**

```python
student = {"Bob":87,"Matt":56,"Sam":29}
```

```python
names = list(student.keys())
values = list(student.values())
```

```python
names
```

```python
values
```

```python
plt.bar(names,values)
plt.show()
```

**Horizontal Bar Plot**

```python
plt.barh(names,values,color='r')
plt.grid(True)
plt.show()
```
